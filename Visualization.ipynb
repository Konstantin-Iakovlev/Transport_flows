```python
import numpy as np
import json
from matplotlib import pyplot as plt

import glob
from scipy.stats import kendalltau, spearmanr, pearsonr
```

```python
for f_name in glob.glob('*_results.json'):
    with open(f_name) as f:
        results = json.load(f)
    plt.figure(figsize=(10, 3))
    gaps = [r['spec_gap'] for r in results.values()]
    mcfp_sol = [r['solution'] for r in results.values()]
    enn = [r['enn_1e-3'] for r in results.values()]
    
    corr, p_val = kendalltau(gaps, mcfp_sol)
    plt.scatter(gaps, mcfp_sol)
    plt.xlabel(r'spectral gap ($\lambda_1 - \lambda_2$)')
    plt.ylabel('exact solution')
    plt.title(f_name + f' kendalltau: {round(corr, 2)}, p_value: {round(p_val, 3)}')
    plt.show()
    
    plt.figure(figsize=(10, 3))
    corr, p_val = kendalltau(enn, mcfp_sol)
    plt.scatter(enn, mcfp_sol)
    plt.xlabel(r'ENN ($\gamma = 10^{-3}$)')
    plt.ylabel('exact solution')
    plt.title(f_name + f' kendalltau: {round(corr, 2)}, p_value: {round(p_val, 3)}')
    plt.show()
    
    
```

### Conclusion

Effective number of neighbours and spectral gap are informative features for the
synthetic data and topology zoo.
